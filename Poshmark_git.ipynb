```python
import warnings
warnings.filterwarnings('error', category=RuntimeWarning)

async def extract_details(link, session): 
    # similar to requests.get but with a different syntax 
    async with session.get(link) as response: 
        # notice that we must await the .text() function  
        soup = BeautifulSoup(await response.text(), 'lxml')
        tag = soup.find(attrs={"class":"timestamp"})
        newtag1 = str(tag).replace('\n', '')
        newtag2 = str(newtag1).replace(' ', '')
#         print('Link', str(links.index(link)))
        date = re.split('>Updated|<', newtag2)[-2]

        # check if item is sold
        if 'SOLD!' in soup.text:
            sold_marker = 1
        else:
            sold_marker = 0

        return date, sold_marker

async def main(links): 
# create an aiohttp session and pass it to each function execution 
    async with aiohttp.ClientSession() as session: 
        tasks = [ 
            extract_details(link, session) 
            for link in links 
        ]
#         results = [await task_ for task_ in tqdm.as_completed(tasks, total=len(tasks))]
        results = await tqdm_asyncio.gather(*tasks) 
        return results
```

```python
# get list of products on page
from bs4 import BeautifulSoup
import re

# empty vectors for variable storage
links = []
product_names = []
prices = []
dates = []
sold_markers = []
        
# update page source
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')

# get soups
product_soup = soup.find_all(attrs={"class":"tile__title tc--b"}, href=True)
price_soup = soup.find_all(attrs={"class":"p--t--1 fw--bold"})
date_soup = soup.find_all("div", {"class": "Row__RowItem-sc-5d290873-0 kEXjGH"})

# get number of sold items, which are stored at the end of the page
num_sold_items = len(soup.find_all(attrs={"class":"icon tile__inventory-tag sold-tag"}))

# get product names and links
for tag in product_soup:
    link = 'https://poshmark.com' + tag.get('href')
    # print(link)
    links.append(link)
    p_name = re.split('\n            |\n', str(tag))[-2]
    product_names.append(p_name)

# get prices
for tag in price_soup:
    newtag1 = str(tag).replace('\n', '')
    newtag2 = str(newtag1).replace(' ', '')
    price = int(re.split('>|<', newtag2)[-3][1:])
    prices.append(price)

# get product dates and sold statuses w/ asyncio
output = await main(links)

# loop through asyncio results and get dates, sold statuses
for t in output:
    dates.append(list(t)[0])
    sold_markers.append(list(t)[1])
    
# item ids
item_ids = [x.split('-')[-1] for x in links]
```

## Drop Price By a Percentage

```python
# SET DISCOUNT AND FLOOR PRICE
percent_discount = 10
floor_price = 10
```

```python
# get rid of recently updated listings
pricedrop_df = df[~df['Last Updated'].str.contains('ago')].reset_index(drop=True).drop_duplicates()

# remove sold listings
pricedrop_df = pricedrop_df.loc[(pricedrop_df['Sold'] == 0) & (pricedrop_df['Price'] > floor_price)].reset_index(drop=True)
```

```python
# price drop loop
j=0
for i in range(0, len(pricedrop_df)):
    item = pricedrop_df.iloc[i,list(pricedrop_df.columns).index('Item')]
    item_id = pricedrop_df.iloc[i,list(pricedrop_df.columns).index('item_id')]
    curr_price = pricedrop_df.iloc[i,list(pricedrop_df.columns).index('Price')]
    reduced_price = round(curr_price - (curr_price * (percent_discount*0.01)))
    # don't drop prices below floor price
    if reduced_price < floor_price:
        reduced_price = floor_price
        
    # don't waste time on products already at the floor price    
    if curr_price == reduced_price:
        print('Item '+ str(i+1) + ' of ' + str(len(pricedrop_df)), '('+item+')\n', 'Old Price: '+ str(curr_price), 'New Price: '+ str(reduced_price))
        print('\tItem already at floor price.')
        
    elif curr_price < reduced_price:
        print('Item '+ str(i+1) + ' of ' + str(len(pricedrop_df)), '('+item+')\n', 'Old Price: '+ str(curr_price), 'New Price: '+ str(reduced_price))
        print('\tItem already priced below floor price.')
        
    else:
        j+=1
        print('Item '+ str(i+1) + ' of ' + str(len(pricedrop_df)), '('+item+')\n' + '\tOld Price: '+ str(curr_price), 'New Price: '+ str(reduced_price) + '\n' + '\t' + 'Memory usage: (%):', str(psutil.virtual_memory().percent))

        # go to edit listing page
        driver.get('https://poshmark.com/edit-listing/' + item_id)

        # sometimes a login pops up when editing, this logs back in for the first listing
        if i == 0 and 'login' in driver.current_url:
            driver.find_element(By.XPATH, '//*[@id="login_form_username_email"]').send_keys(username)
            driver.find_element(By.XPATH, '//*[@id="login_form_password"]').send_keys(password)
            driver.find_element(By.XPATH, '//*[@id="login_form_password"]').send_keys(Keys.RETURN)

        time.sleep(5) # wait for page to load
        page = driver.find_element("tag name","html")
        page.send_keys(Keys.PAGE_DOWN)
        page.send_keys(Keys.PAGE_DOWN)
        page.send_keys(Keys.PAGE_DOWN)
        page.send_keys(Keys.PAGE_DOWN)
        price_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/main/div[2]/div/div[1]/div/section[9]/div/div/div[2]/div[1]/div[3]/input')))
        price_element.click()

        # clear current price
        for i in range(0,4):
            price_element.send_keys(Keys.BACKSPACE)

        price_element.send_keys(str(reduced_price))
        price_element.send_keys(Keys.RETURN)
        time.sleep(1)
        page.send_keys(Keys.END)
        update_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/main/div[2]/div/div[1]/div/div[2]/button')))
        update_button.click()
        time.sleep(1)
        list_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@class="bg--light-gray p--3 ta--c"]')))
        list_button.click()
        time.sleep(1)
        # handle fragile item button if it pops up
        try:
            fragile_warning_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/div/div[1]/div/div[6]/div/div[1]/div[2]/div[3]/div/button[2]')))
            fragile_warning_button.click()
        except:
            pass
        
    
print('Done.')
driver.close()
```
