# Get Current Products

```python
# go to facebook marketplace listings
driver.get('https://www.facebook.com/marketplace/you/selling')
```

```python
# grid view
driver.find_element(By.XPATH, "//*[@aria-label='Grid View']").click()
```

```python
# scroll down to bottom while allowing time for products to load V3
ypos_list = []

try:
    curr_ypos = round(driver.find_element(By.XPATH,  "//*[@aria-label='Loading...']").rect['y'], 2)
except:
    pass

for i in range(0,1000):
    page = driver.find_element("tag name","html")
    try:
        rect_dict = driver.find_element(By.XPATH,  "//*[@aria-label='Loading...']").rect
    except:
        break
    loading_ypos = round(rect_dict['y'], 2)
    page.send_keys(Keys.END)
    time.sleep(0.5)
    if len(ypos_list) == 15:
        ypos_list.pop(0)
    ypos_list.append(loading_ypos)
    
    # if the past 15 y positions have been the same, stop scrolling
    if len(ypos_list) == 15 and all([x == loading_ypos for x in ypos_list]):
        page.send_keys(Keys.END)
        break
    curr_ypos = loading_ypos
```

```python
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
```

```python
# get current year
today = datetime.datetime.now()
curr_year = today.year

product_names = []
prices = []
dates = []
sold_statuses = []

product_soup = soup.find_all("div",attrs={"class":"x1lq5wgf xgqcy7u x30kzoy x9jhf4c x1h5jrl4 xw7yly9 xh8yej3"})
for tag in product_soup:
#     print(product_soup.index(tag))
    if "Tip:" in tag.text:
        prod_str = tag.text.replace('Tip: Renew your listing?', '')
    else:
        prod_str = tag.text
    if 'Active' in prod_str:
        status = 'Active'
        sold_statuses.append(status)
         # determine whether there are 1 or 2 prices in the string
        if prod_str.count('$') < 2:
            prod_name, price, garbage, date = re.split('\\$|Active|Listed on ',prod_str)
            # count forward slashes in date, add year suffix if < 2 forward slashes
            if date.count('/') < 2:
                date = date + '/' + str(curr_year)
            product_names.append(prod_name)
            prices.append(int(price))
            dates.append(date)
        else:
            prod_name, price, oldprice, garbage, date = re.split('\\$|Active|Listed on ',prod_str)
            # count forward slashes in date, add year suffix if < 2 forward slashes
            if date.count('/') < 2:
                date = date + '/' + str(curr_year)
            product_names.append(prod_name)
            prices.append(int(price))
            dates.append(date)
#         print(prod_name, price, date)
    elif 'Sold' in prod_str:
        status = 'Sold'
        sold_statuses.append(status)
        # determine whether there are 1 or 2 prices in the string
        if prod_str.count('$') < 2:
            prod_name, price, garbage, date = re.split('\\$|Sold|Listed on ',prod_str)
            if date.count('/') < 2:
                date = date + '/' + str(curr_year)
            product_names.append(prod_name)
            prices.append(int(price))
            dates.append(date)
        else:
            prod_name, price, oldprice, garbage, date = re.split('\\$|Sold|Listed on ',prod_str)
            if date.count('/') < 2:
                date = date + '/' + str(curr_year)
            product_names.append(prod_name)
            prices.append(int(price))
            dates.append(date)
```

```python
# dataframe w/ prices
dict = {'Item': product_names, 'Price': prices, 'Listed Date': dates, 'Sold Status' : sold_statuses} 
df = pd.DataFrame(dict)

# convert Listed Date column to datetime format
df['Listed Date'] = pd.to_datetime(df['Listed Date'], format='%m/%d/%Y')
timestamps = today - df['Listed Date']

# get days from timestamps
days_since = []
for a in timestamps:
    days_since.append(a.days)

df['Days Since Listing'] = days_since
    
display(df)
```

```python
# from https://stackoverflow.com/questions/44777053/selenium-movetargetoutofboundsexception-with-firefox
def scroll_shim(passed_in_driver, object):
        x = object.location['x']
        y = object.location['y']
        scroll_by_coord = 'window.scrollTo(%s,%s);' % (
            x,
            y
        )
        scroll_nav_out_of_way = 'window.scrollBy(0, -120);'
        passed_in_driver.execute_script(scroll_by_coord)
        passed_in_driver.execute_script(scroll_nav_out_of_way)
```

```python
# loop through "Renew listing" buttons and click them
for button in driver.find_elements(By.XPATH, '//span[text()="Renew listing"]'):
    try:
        if 'firefox' in driver.capabilities['browserName']:
            scroll_shim(driver, button)
        # scroll_shim is just scrolling it into view, still need to hover over it to click using an action chain.
        actions = ActionChains(driver)
        actions.move_to_element(button)
        actions.click()
        actions.perform()
    except:
        pass
```
